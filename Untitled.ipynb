```python
task_user_count = data[data.Language=='C'].groupby(['QCode','UserID']).size()
```

```python
# Atleast 2 submissions
df_trgt = pd.DataFrame(task_user_count[task_user_count>=2])
```

```python
# number of users for every problem with atleast 2 submissions
df_trgt.reset_index().QCode.value_counts()
```

```python
# Distribution of Number of users with atleast 2 submissions
df_trgt.reset_index().QCode.value_counts().describe(percentiles=[.25,.3,.4,.5,.6,.75,.80,.90,.99])
```
